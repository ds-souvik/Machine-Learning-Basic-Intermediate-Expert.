```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[2,5,7])
plt.show()
```

```python
plt.plot([1,2,3],[2,5,7])
plt.title('Heading of Graph')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.show()
```

```python
X=[1,2,30]
Y=[3,5,9]
X2=[34,22,43]
Y2=[2,33,66]
plt.plot(X,Y)
plt.plot(X2, Y2)
plt.show
```

```python
from matplotlib import style
style.use('ggplot')
X=[1,2,30]
Y=[3,5,9]
X2=[34,22,43]
Y2=[2,33,66]
plt.plot(X,Y,label="First graph")
plt.plot(X2, Y2,label="Second graph")
plt.legend()
plt.show
```
