### 1.3 Identifying outliers and treating them.

```python
#Getting the numerical and categorical columns
num_col = leads.select_dtypes(include=np.number).columns

cat_col = leads.select_dtypes(include=np.object).columns
(num_col,cat_col)
```

```python
# Checking outliers for numerical columns excluding the target using boxplots:
plt.figure(figsize = (16,8))
subplot = 221
for i in leads[num_col.drop('Converted')]:
    plt.subplot(subplot)
    sns.boxplot(leads[i])
    subplot +=1
plt.tight_layout()
```

#### Based on the boxplots and describe function on the dataset:
1. The average Total Visits by a prospect customer and lead procurred was 3, minimum was 0 and 75% was 5 while the maximum Total Visits is 251 which is clearly an outlier. Also there are other outliers which are clearly visible from the boxplots.
2. There are no visible outliers for Total Time Spend on Website. Need to analyze more on that.
3. The average Page Views Per Visit by a prospect customer and lead procurred was 2, minimum was 0 and 75% was 3 while the maximum Page Views Per Visit is 55 which is clearly an outlier. Also there are other outliers which are clearly visible from the boxplots.

```python
def num_outliers(dataframe):
    q1=dataframe.quantile(0.25)
    q3=dataframe.quantile(0.75)
    iqr=q3-q1
    outlier=((dataframe < (q1 - 1.5 * iqr)) | (dataframe > (q3 + 1.5 * iqr))).sum()
    print('q1')
    print(q1)
    print('\nq3')
    print(q3)
    print('\nIQR ')
    print(iqr)
    print('\nNumber of outliers')
    return outlier
outlier=num_outliers(leads[num_col.drop('Converted')])
outlier
```

#### Statistically there are no outliers on Total Time Spent on Website.
#### Based on experience for columns TotalVisits and Page Views Per Visit, values > 1.5 * IQR might not serve as the perfect outliers that requires treatment.
#### For TotalVisits value>30  and for Page Views Per Visit > 15 can serve as  <font color='red'>major outliers.</font>
Further analyzing the value counts for these outliers

```python
#Checking number of datapoints with our assumed outliers along with our target variable distribution.
print(leads[leads['TotalVisits'] > 30].Converted.value_counts())
print(leads[leads['Page Views Per Visit'] > 15].Converted.value_counts())
```

```python
#Outlier treatment
leads['TotalVisits']=leads['TotalVisits'].apply(lambda x: 30 if x>30 else x)
leads['Page Views Per Visit']=leads['Page Views Per Visit'].apply(lambda x: 15 if x>15 else x)
```

# 3. Data Preparation.

## 3.1 Dummy variable creation

```python
pd.set_option('display.max_columns', None)
leads.head()
```

```python
#Keeping the dataframe leads as the one without any binary mapping or dummy variables

leads_computed = leads.copy()

# Converting the binary variables (Yes/No) to 0/1

# List of variables to map
varlist =  ['Do Not Email', 'A free copy of Mastering The Interview']

# Defining the map function
def binary_map(x):
    return x.map({'Yes': 1, "No": 0})

# Applying the function to the leads list
leads_computed[varlist] = leads_computed[varlist].apply(binary_map)
```

```python
leads_computed.head()
```

```python
leads_computed.columns
```

```python
#Creating Dummy Variables.

# Creating dummy variables for the variable 'Lead Origin'
ml = pd.get_dummies(leads_computed['Lead Origin'], prefix='Lead Origin')
# Dropping MultipleLines_No phone service column
ml1 = ml.drop(['Lead Origin_Quick Add Form'], 1)
#Adding the results to the master dataframe
leads_computed = pd.concat([leads_computed,ml1], axis=1)

# Creating dummy variables for the variable 'Lead Source'
ml = pd.get_dummies(leads_computed['Lead Source'], prefix='Lead Source')
# Dropping MultipleLines_No phone service column
ml1 = ml.drop(['Lead Source_NC_EDM'], 1)
#Adding the results to the master dataframe
leads_computed = pd.concat([leads_computed,ml1], axis=1)

# Creating dummy variables for the variable 'Last Activity'
ml = pd.get_dummies(leads_computed['Last Activity'], prefix='Last Activity')
# Dropping MultipleLines_No phone service column
ml1 = ml.drop(['Last Activity_Resubscribed to emails'], 1)
#Adding the results to the master dataframe
leads_computed = pd.concat([leads_computed,ml1], axis=1)

# # Creating dummy variables for the variable 'Country'
# ml = pd.get_dummies(leads_computed['Country'], prefix='Country')
# # Dropping MultipleLines_No phone service column
# ml1 = ml.drop(['Country_Indonesia'], 1)
# #Adding the results to the master dataframe
# leads_computed = pd.concat([leads_computed,ml1], axis=1)

# Creating dummy variables for the variable 'Specialization'
ml = pd.get_dummies(leads_computed['Specialization'], prefix='Specialization')
# Dropping MultipleLines_No phone service column
ml1 = ml.drop(['Specialization_E-Business'], 1)
#Adding the results to the master dataframe
leads_computed = pd.concat([leads_computed,ml1], axis=1)

# Creating dummy variables for the variable 'What is your current occupation'
ml = pd.get_dummies(leads_computed['What is your current occupation'], prefix='What is your current occupation')
# Dropping MultipleLines_No phone service column
ml1 = ml.drop(['What is your current occupation_Housewife'], 1)
#Adding the results to the master dataframe
leads_computed = pd.concat([leads_computed,ml1], axis=1)


# Creating dummy variables for the variable 'Tags'
ml = pd.get_dummies(leads_computed['Tags'], prefix='Tags')
# Dropping MultipleLines_No phone service column
ml1 = ml.drop(['Tags_University not recognized'], 1)
#Adding the results to the master dataframe
leads_computed = pd.concat([leads_computed,ml1], axis=1)

# Creating dummy variables for the variable 'City'
ml = pd.get_dummies(leads_computed['City'], prefix='City')
# Dropping MultipleLines_No phone service column
ml1 = ml.drop(['City_Tier II Cities'], 1)
#Adding the results to the master dataframe
leads_computed = pd.concat([leads_computed,ml1], axis=1)

# Creating dummy variables for the variable 'Last Notable Activity'
ml = pd.get_dummies(leads_computed['Last Notable Activity'], prefix='Last Notable Activity')
# Dropping MultipleLines_No phone service column
ml1 = ml.drop(['Last Notable Activity_Email Marked Spam'], 1)
#Adding the results to the master dataframe
leads_computed = pd.concat([leads_computed,ml1], axis=1)
```

```python
#Dropping the columns for which dummy variables are created.

leads_computed=leads_computed.drop(['Lead Origin', 'Lead Source', 'Last Activity', 'Specialization', 'What is your current occupation',
                    'Tags', 'City', 'Last Notable Activity'], axis=1)
leads_computed.head()
```

## 3.2 Train test split

```python
# Splitting data into X and y
X=leads_computed.drop('Converted', axis=1)
y=leads_computed['Converted']
print(X.shape, y.shape)

# Splitting X and y into training and test datasets
X_train, X_test, y_train, y_test= train_test_split(X, y, random_state=100, train_size=0.7)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```
