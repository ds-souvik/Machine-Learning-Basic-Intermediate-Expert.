Now there's no literature available which states the exact rule of thumb specifying the percentage of acceptable null values.
But market standard says most of the times we can drop columns with more than 25-30% missing data, of-course with many exceptions as per the use case.

Also Here Occupation type lies in the border line with 31% null values. But we believe Occupation type plays a significant role to explore if it can be used to find whether the person will default or not. Hence we are deciding to drop columns columns with more than 40% of missing values(Since there's no value between 31-40%)

```python
#Finding the number of columns having more than 30% rows with null values
len(app.isnull().sum()[app.isnull().sum().values>(0.40 * len(app))])
```

#### Listing down these 49 columns that are to be dropped:

BC itna ghatiya data maine apne zindagi me kabhi nai dekha.

```python
insignificant_col1 = list(app.isnull().sum()[app.isnull().sum().values>(0.40 * len(app))].index)
print(insignificant_col1)
print(len(insignificant_col1))
```

```python
#Dropping these columns
app.drop(labels=insignificant_col1, axis=1, inplace=True)
app.shape
```

```python
#Rest of the remaining columns having null values
app.isnull().sum()[app.isnull().sum()>0]
```

Not getting the meaning of AMT_ANNUITY. Since it is only 12 columns we can impute it.

```python
print("AMT_ANNUITY")
print('Mean: ', app['AMT_ANNUITY'].mean())
print('Median: ', app['AMT_ANNUITY'].median())
print('Mode: ', app['AMT_ANNUITY'].mode())

print("----------------------------------")
print("AMT_GOODS_PRICE")
print('Mean: ', app['AMT_GOODS_PRICE'].mean())
print('Median: ', app['AMT_GOODS_PRICE'].median())
print('Mode: ', app['AMT_GOODS_PRICE'].mode())
```

```python
app[app['AMT_ANNUITY'].isnull()]
```

Looking at the AMT_CREDIT for rows with AMT_ANNUITY null values, replacing null values with mode doesn't seem to be a good idea. Hence replacing the null values with median.

```python
app['AMT_ANNUITY'].fillna((app['AMT_ANNUITY'].median()), inplace=True)
```

```python
app[app['AMT_GOODS_PRICE'].isnull()]
```

Null values of AMT_GOODS_PRICE will have to be imputed using mode.

```python
app['AMT_GOODS_PRICE'].fillna((app['AMT_GOODS_PRICE'].median()), inplace=True)
```

There are still columns with more than 40000 null values. On examining those columns, they won't be of any significance further in our analysis. Dropping those columns.

```python
insignificant_col2=list(app.isnull().sum()[app.isnull().sum().values>40000].index)
insignificant_col2
app.drop(labels=insignificant_col2, axis=1, inplace=True)
app.isnull().sum()[app.isnull().sum().values>0]
```

On further examination, the columns NAME_TYPE_SUITE and EXT_SOURCE_2 seemed insignificant. Hence dropping those columns.

```python
insignificant_col3=['NAME_TYPE_SUITE', 'EXT_SOURCE_2']
app.drop(labels=insignificant_col3, axis=1, inplace=True)
app.isnull().sum()[app.isnull().sum().values>0]
```

Imputing the 2 missing values for CNT_FAM_MEMBERS with mode since mean can give incorrect values and median will give the value in center. Since only 2 rows are impacted going for the most occured value.

Dropping the row which has missing value in column DAYS_LAST_PHONE_CHANGE .

```python
app['CNT_FAM_MEMBERS'].fillna((app['CNT_FAM_MEMBERS'].mode()), inplace=True)
```
