## Writing a csv file

```python
df1=pd.DataFrame([[1,3,2,4],[5,3,6,8],[11,5,4,33],[4,7,88,3]],columns=['A','B','C','D'])
df1
```

```python
df1.to_csv('My First Dataframe.csv')
```

```python
#When the volume of data is more it's always a good practice to  have index=False to save disk space.
df1.to_csv('My Second Dataframe.csv',index=False)
```

## Merging Dataframes

```python
df1
```

```python
df2=pd.DataFrame([[11,3,22,44],[55,3,66,88],[111,5,44,333],[44,7,888,33]],columns=['W','X','Y','Z'])
```

```python
df2
```

```python
df3=pd.merge(df1,df2,right_on='X',left_on='B')
```
