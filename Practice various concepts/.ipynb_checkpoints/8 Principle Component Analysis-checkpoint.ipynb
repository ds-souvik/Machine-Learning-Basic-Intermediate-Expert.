# `Principal Component Analysis on Breast Cancer Dataset`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import sklearn
```

```python
#Import dataset
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer=load_breast_cancer()
```

```python
#Create dataframe from imported data
df=pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
#targets
breast_cancer.target_names
```

```python
#Checking for null values
df.isnull().sum()
```

```python
df.shape
```

```python
#Scaling the variables.
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df)
scaled_data=scaler.transform(df)
```

```python
scaled_data_df=pd.DataFrame(scaled_data, columns=breast_cancer.feature_names)
```

```python
plt.figure(figsize=(20,18))
sns.set_palette("husl")
sns.heatmap(scaled_data_df.corr(),annot=True)
plt.show()
```

```python
scaled_data_df
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
sns.distplot(scaled_data_df['mean radius'])
plt.subplot(1,3,2)
sns.distplot(scaled_data_df['smoothness error'])
plt.subplot(1,3,3)
sns.distplot(scaled_data_df['worst perimeter'])
plt.show()
```

```python
#Distribution kind of has become Gaussian
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(scaled_data_df)
```

```python
x_pca=pca.transform(scaled_data_df)
```

```python
x_pca.shape
```

```python
#Reduced 30 dimensions to 2
```

```python
x_pca_df=pd.DataFrame(x_pca, columns=['Dimension_1', 'Dimension_2'])
x_pca_df.head()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=x_pca_df['Dimension_1'], y=x_pca_df['Dimension_2'], c=breast_cancer.target, cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
plt.show()
```
