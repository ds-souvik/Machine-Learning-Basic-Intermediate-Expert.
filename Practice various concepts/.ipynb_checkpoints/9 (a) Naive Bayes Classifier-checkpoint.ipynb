# `Naive Bayes Algorithm`

```python
#Impoprting Libraries
import numpy as np
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import load_breast_cancer
```

```python
#Loading the dataset
breast_cancer=load_breast_cancer()
```

```python
#Checking the target names
breast_cancer.target_names
```

```python
#Creating dataframe out of data
df=pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
df['target']=breast_cancer.target
```

```python
#Code to show all columns
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
#Checking null values
df.isnull().sum()
```

```python
#Identifying Multi-collinear variables
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif=pd.DataFrame()
vif['Features']=df.columns
```

```python
for column in df.columns:
    plt.figure(figsize=(20,5))
    plt.subplot(121)
    sns.distplot(df[column])
    plt.title(column)
```

```python
vif['VIF']=[variance_inflation_factor(df.values,i) for i in range(df.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif=vif.sort_values(by='VIF', ascending=False)
vif
```

```python
#Dividing the dataset into indepenent and dependent variables
y=df.pop('target')
X=df
```

```python
#Splitting the train test dataset for independent and dependent variable
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, random_state=42)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Initializing and fiting Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(X_train, y_train)
```

```python
#Checking accuracy
gnb.score(X_test, y_test)
```

```python
#Prediction
y_test_pred=gnb.predict(X_test)
y_test_pred
```

```python
#Importing libraries for accuracy_score and confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
#accuracy_score
accuracy_score(y_test, y_test_pred)
```

```python
confusion=confusion_matrix(y_test, y_test_pred)
```

```python
#Sensitivity and Specificity
Sensitivity=confusion[0,0]/(confusion[0,0]+confusion[0,1])
Specificity=confusion[1,1]/(confusion[1,0]+confusion[1,1])
print(Sensitivity)
print(Specificity)
```
